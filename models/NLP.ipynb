```python

import torch
import torch.nn as nn
import torch.nn.functional as F
import math
from dataclasses import dataclass, field
from typing import List, Optional

@dataclass
class config:
    vocab_size: int
    encoder_block_size: int
    decoder_block_size: int
    n_layer: int
    n_head: int
    n_embd: int
    dropout: float
    pad_token_id: int
    special_token_ids: List[int] = field(default_factory=list)


class SelfAttention(nn.Module):
    def __init__(self, config: config, is_causal: bool = True, block_size: int = None):
        super().__init__()
        assert config.n_embd % config.n_head == 0
        self.n_head = config.n_head
        self.n_embd = config.n_embd
        self.c_attn = nn.Linear(config.n_embd, 3 * config.n_embd, bias=False)
        self.attn_drop = nn.Dropout(config.dropout)
        self.is_causal = is_causal

        if self.is_causal:
            self.register_buffer(
                "bias",
                torch.tril(torch.ones(block_size, block_size))
                .view(1, 1, block_size, block_size)
            )

        self.c_proj = nn.Linear(config.n_embd, config.n_embd)

    def forward(self, x, pad_mask=None):
        B, T, C = x.size()
        head_dim = C // self.n_head

        qkv = self.c_attn(x)
        q, k, v = qkv.split(C, dim=2)

        q = q.view(B, T, self.n_head, head_dim).transpose(1, 2)
        k = k.view(B, T, self.n_head, head_dim).transpose(1, 2)
        v = v.view(B, T, self.n_head, head_dim).transpose(1, 2)

        att = (q @ k.transpose(-2, -1)) * (1.0 / math.sqrt(head_dim))

        if self.is_causal:
            att = att.masked_fill(self.bias[:, :, :T, :T] == 0, float("-inf"))

        if pad_mask is not None:
            att = att.masked_fill((pad_mask[:, None, None, :T] == 0), float("-inf"))

        att = F.softmax(att, dim=-1)
        att = self.attn_drop(att)

        y = att @ v
        y = y.transpose(1, 2).contiguous().view(B, T, C)
        y = self.c_proj(y)
        return y


class MLP(nn.Module):
    def __init__(self, config: config):
        super().__init__()
        self.fc   = nn.Linear(config.n_embd, 4 * config.n_embd)
        self.proj = nn.Linear(4 * config.n_embd, config.n_embd)
        self.drop = nn.Dropout(config.dropout)

    def forward(self, x):
        return self.drop(self.proj(F.gelu(self.fc(x))))


class Block(nn.Module):
    def __init__(self, config: config, is_causal: bool = True, block_size: int = None):
        super().__init__()
        self.ln_1 = nn.LayerNorm(config.n_embd)
        self.attn = SelfAttention(config, is_causal=is_causal, block_size=block_size)
        self.ln_2 = nn.LayerNorm(config.n_embd)
        self.mlp  = MLP(config)

    def forward(self, x, pad_mask=None):
        x = x + self.attn(self.ln_1(x), pad_mask=pad_mask)
        x = x + self.mlp(self.ln_2(x))
        return x


class CrossAttention(nn.Module):
    def __init__(self, config: config):
        super().__init__()
        assert config.n_embd % config.n_head == 0
        self.n_head = config.n_head
        self.n_embd = config.n_embd

        self.query_proj = nn.Linear(config.n_embd, config.n_embd, bias=False)
        self.key_proj   = nn.Linear(config.n_embd, config.n_embd, bias=False)
        self.value_proj = nn.Linear(config.n_embd, config.n_embd, bias=False)

        self.attn_drop = nn.Dropout(config.dropout)
        self.c_proj    = nn.Linear(config.n_embd, config.n_embd)

    def forward(self, decoder_hidden_states, encoder_output, encoder_pad_mask=None):
        B, T_dec, C = decoder_hidden_states.size()
        _, T_enc, _ = encoder_output.size()
        head_dim = C // self.n_head

        q = self.query_proj(decoder_hidden_states)
        k = self.key_proj(encoder_output)
        v = self.value_proj(encoder_output)

        q = q.view(B, T_dec, self.n_head, head_dim).transpose(1, 2)
        k = k.view(B, T_enc, self.n_head, head_dim).transpose(1, 2)
        v = v.view(B, T_enc, self.n_head, head_dim).transpose(1, 2)

        att = (q @ k.transpose(-2, -1)) * (1.0 / math.sqrt(head_dim))

        if encoder_pad_mask is not None:
            att = att.masked_fill(
                (encoder_pad_mask[:, None, None, :] == 0), float("-inf")
            )

        att = F.softmax(att, dim=-1)
        att = self.attn_drop(att)

        y = att @ v
        y = y.transpose(1, 2).contiguous().view(B, T_dec, C)
        y = self.c_proj(y)
        return y, att


class DecoderBlock(nn.Module):
    def __init__(self, config: config):
        super().__init__()
        self.ln_1      = nn.LayerNorm(config.n_embd)
        self.self_attn = SelfAttention(config, is_causal=True,
                                       block_size=config.decoder_block_size)
        self.ln_2       = nn.LayerNorm(config.n_embd)
        self.cross_attn = CrossAttention(config)
        self.ln_3       = nn.LayerNorm(config.n_embd)
        self.mlp        = MLP(config)

    def forward(self, x, encoder_output, decoder_pad_mask=None, encoder_pad_mask=None):
        x = x + self.self_attn(self.ln_1(x), pad_mask=decoder_pad_mask)
        y, cross_attn_weights = self.cross_attn(
            self.ln_2(x), encoder_output, encoder_pad_mask=encoder_pad_mask
        )
        x = x + y
        x = x + self.mlp(self.ln_3(x))
        return x, cross_attn_weights


class Encoder(nn.Module):
    def __init__(self, config: config):
        super().__init__()
        self.config = config
        self.wte  = nn.Embedding(config.vocab_size, config.n_embd)
        self.wpe  = nn.Embedding(config.encoder_block_size, config.n_embd)
        self.drop = nn.Dropout(config.dropout)
        self.h    = nn.ModuleList([
            Block(config, is_causal=False, block_size=config.encoder_block_size)
            for _ in range(config.n_layer)
        ])
        self.ln_f = nn.LayerNorm(config.n_embd)
        self.apply(self._init_weights)

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            nn.init.normal_(module.weight, mean=0.0, std=0.02)
            if module.bias is not None:
                nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            nn.init.normal_(module.weight, mean=0.0, std=0.02)

    def forward(self, input_ids, pad_mask=None):
        B, T = input_ids.size()
        assert T <= self.config.encoder_block_size

        pos     = torch.arange(T, device=input_ids.device).unsqueeze(0)
        x       = self.drop(self.wte(input_ids) + self.wpe(pos))
        for block in self.h:
            x = block(x, pad_mask=pad_mask)
        return self.ln_f(x)


class Decoder(nn.Module):
    def __init__(self, config: config):
        super().__init__()
        self.config = config
        self.wte  = nn.Embedding(config.vocab_size, config.n_embd)
        self.wpe  = nn.Embedding(config.decoder_block_size, config.n_embd)
        self.drop = nn.Dropout(config.dropout)
        self.h    = nn.ModuleList([DecoderBlock(config) for _ in range(config.n_layer)])
        self.ln_f = nn.LayerNorm(config.n_embd)
        self.lm_head = nn.Linear(config.n_embd, config.vocab_size, bias=False)
        self.lm_head.weight = self.wte.weight          # weight tying
        self.p_gen = nn.Linear(config.n_embd, 1)
        self.apply(self._init_weights)

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            nn.init.normal_(module.weight, mean=0.0, std=0.02)
            if module.bias is not None:
                nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            nn.init.normal_(module.weight, mean=0.0, std=0.02)

    def forward(self, decoder_input_ids, encoder_output,
                decoder_pad_mask=None, encoder_pad_mask=None):
        B, T = decoder_input_ids.size()
        assert T <= self.config.decoder_block_size

        pos = torch.arange(T, device=decoder_input_ids.device).unsqueeze(0)
        x   = self.drop(self.wte(decoder_input_ids) + self.wpe(pos))

        attn_weight_list = []
        for block in self.h:
            x, attn = block(x, encoder_output,
                            decoder_pad_mask=decoder_pad_mask,
                            encoder_pad_mask=encoder_pad_mask)
            attn_weight_list.append(attn)

        x = self.ln_f(x)

        p_gen       = torch.sigmoid(self.p_gen(x))                        # (B, T, 1)
        final_attn  = sum(attn_weight_list) / len(attn_weight_list)        # avg over layers
        final_attn  = final_attn.mean(dim=1)                               # avg over heads → (B, T_dec, T_enc)
        logits      = self.lm_head(x)

        return logits, final_attn, p_gen


class Seq2SeqModel(nn.Module):

    def __init__(self, config: config):
        super().__init__()
        self.config       = config
        self.pad_token_id = config.pad_token_id
        #build a set of token ids that must never appear in the copy distribution
        self._no_copy_ids = set(config.special_token_ids) | {config.pad_token_id}

        self.encoder = Encoder(config)
        self.decoder = Decoder(config)
        self.apply(self._init_weights)

    def _init_weights(self, module):
        if isinstance(module, nn.Linear):
            nn.init.normal_(module.weight, mean=0.0, std=0.02)
            if module.bias is not None:
                nn.init.zeros_(module.bias)
        elif isinstance(module, nn.Embedding):
            nn.init.normal_(module.weight, mean=0.0, std=0.02)

    def _compute_copy_logits(self, attn_weights, encoder_input_ids):
        B, T_dec, T_enc = attn_weights.shape
        copy_dist = torch.zeros(B, T_dec, self.config.vocab_size,
                                device=attn_weights.device, dtype=attn_weights.dtype)

        # Build mask: True = token is allowed to be copied
        copy_mask = torch.ones_like(encoder_input_ids, dtype=torch.bool)
        for sid in self._no_copy_ids:
            copy_mask &= (encoder_input_ids != sid)

        for b in range(B):
            valid_idx      = torch.where(copy_mask[b])[0]
            if valid_idx.numel() == 0:
                continue
            valid_token_ids = encoder_input_ids[b, valid_idx]
            valid_attn      = attn_weights[b, :, valid_idx]

            index = valid_token_ids.unsqueeze(0).expand(T_dec, -1)
            copy_dist[b].scatter_add_(1, index, valid_attn)

        #renormalise so copy side always sums to 1
        copy_sum = copy_dist.sum(dim=-1, keepdim=True).clamp(min=1e-10)
        copy_dist = copy_dist / copy_sum
        return copy_dist

    def forward(self, encoder_input_ids, decoder_input_ids,
                targets=None, use_scheduled_sampling=False, sampling_prob=0.0):

        encoder_pad_mask = (encoder_input_ids != self.pad_token_id).float()
        encoder_output   = self.encoder(encoder_input_ids, pad_mask=encoder_pad_mask)

        # pass encoder_input_ids so scheduled sampling can use the full
        #pointer-network probability (not just generation logits)
        if use_scheduled_sampling and self.training and sampling_prob > 0:
            decoder_input_ids = self._scheduled_sampling(
                decoder_input_ids, encoder_output,
                encoder_pad_mask, encoder_input_ids, sampling_prob
            )

        decoder_pad_mask = (decoder_input_ids != self.pad_token_id).float()
        logits, attn_weights, p_gen = self.decoder(
            decoder_input_ids, encoder_output,
            decoder_pad_mask=decoder_pad_mask,
            encoder_pad_mask=encoder_pad_mask
        )

        gen_probs   = F.softmax(logits, dim=-1)
        copy_probs  = self._compute_copy_logits(attn_weights, encoder_input_ids)
        final_probs = p_gen * gen_probs + (1 - p_gen) * copy_probs
        final_log_probs = torch.log(final_probs + 1e-10)

        nll_loss            = torch.tensor(0.0, device=logits.device)
        attn_entropy_val    = 0.0
        mean_cross_attn_val = 0.0
        loss                = None

        if targets is not None:
            nll_loss = F.nll_loss(
                final_log_probs.view(-1, self.config.vocab_size),
                targets.view(-1),
                ignore_index=-100,
                reduction='mean'
            )

            valid_attn = attn_weights * encoder_pad_mask.unsqueeze(1)

            attn_entropy = -torch.sum(
                valid_attn * torch.log(valid_attn + 1e-10), dim=-1
            )
            attn_entropy_val = attn_entropy.mean().item()

            max_acceptable_entropy = 3.0
            attention_penalty = F.relu(attn_entropy.mean() - max_acceptable_entropy)

            mean_cross_attn     = valid_attn.sum(dim=-1).mean()
            mean_cross_attn_val = mean_cross_attn.item()

            min_attn_threshold  = 0.5
            attention_usage_penalty = F.relu(min_attn_threshold - mean_cross_attn)

            # Vocabulary diversity penalty
            with torch.no_grad():
                pred_tokens = final_probs.argmax(dim=-1)

            vocab_diversity_losses = []
            for b in range(pred_tokens.size(0)):
                seq       = pred_tokens[b]
                valid_seq = seq[targets[b] != -100]
                if len(valid_seq) > 1:
                    unique_ratio  = len(torch.unique(valid_seq)) / len(valid_seq)
                    diversity_loss = (0.5 - unique_ratio) ** 2
                    vocab_diversity_losses.append(diversity_loss)

            if vocab_diversity_losses:
                vocab_diversity_penalty = torch.tensor(
                    vocab_diversity_losses, device=logits.device
                ).mean()
            else:
                vocab_diversity_penalty = torch.tensor(0.0, device=logits.device)

            gen_encouragement = torch.mean((1 - p_gen) ** 2)

            # shift cumsum by 1 along the decoder dimension so coverage_t = sum_{s<t} attn_s
            cumulative      = torch.cumsum(attn_weights, dim=1)
            prev_coverage   = F.pad(cumulative[:, :-1, :], (0, 0, 1, 0))   # shift right by 1
            coverage_loss   = torch.sum(
                torch.min(attn_weights, prev_coverage), dim=-1
            ).mean()

            loss = (
                nll_loss
                + 0.3  * attention_penalty
                + 0.2  * attention_usage_penalty
                + 0.1  * vocab_diversity_penalty
                + 0.15 * gen_encouragement #updated for based of of v0.1 of current model it copys a bit too much
                + 0.1  * coverage_loss
            )

        return final_log_probs, loss, {
            'nll_loss':        nll_loss.item(),
            'attn_entropy':    attn_entropy_val,
            'mean_cross_attn': mean_cross_attn_val,
            'p_gen_mean':      p_gen.mean().item(),
        }

    # ── Scheduled sampling ────────────────────────────────────────────────────
    def _scheduled_sampling(self, decoder_input_ids, encoder_output,
                            encoder_pad_mask, encoder_input_ids, sampling_prob):
        B, T          = decoder_input_ids.size()
        new_dec_input = decoder_input_ids.clone()

        with torch.no_grad():
            for t in range(1, T):
                if torch.rand(1).item() >= sampling_prob:
                    continue

                partial_input    = new_dec_input[:, :t]
                decoder_pad_mask = torch.ones(
                    (B, t), dtype=torch.float, device=decoder_input_ids.device
                )

                logits, attn_weights, p_gen = self.decoder(
                    partial_input, encoder_output,
                    decoder_pad_mask=decoder_pad_mask,
                    encoder_pad_mask=encoder_pad_mask
                )

                # Use only the last decoder position
                last_gen  = F.softmax(logits[:, -1:, :], dim=-1)
                last_attn = attn_weights[:, -1:, :]
                last_p    = p_gen[:, -1:, :]

                last_copy  = self._compute_copy_logits(last_attn, encoder_input_ids)
                last_probs = (last_p * last_gen + (1 - last_p) * last_copy)[:, 0, :]

                last_probs = last_probs.clamp(min=0)
                last_probs = last_probs / last_probs.sum(dim=-1, keepdim=True).clamp(min=1e-10)

                sampled = torch.multinomial(last_probs, num_samples=1)
                new_dec_input[:, t] = sampled.squeeze(-1)

        return new_dec_input

    @torch.no_grad()
    def summarize(self, encoder_input_ids, max_new_tokens,temperature=1.0, top_k=50,bos_token_id=None, eos_token_id=None,repetition_penalty=1.5):
        if bos_token_id is None:
            raise ValueError("bos_token_id must be provided")

        self.eval()
        encoder_pad_mask = (encoder_input_ids != self.pad_token_id).float()
        encoder_output   = self.encoder(encoder_input_ids, pad_mask=encoder_pad_mask)

        decoder_input_ids = torch.tensor([[bos_token_id]], device=encoder_input_ids.device)

        for _ in range(max_new_tokens):
            cur_dec = decoder_input_ids
            if cur_dec.size(1) > self.config.decoder_block_size:
                cur_dec = cur_dec[:, -self.config.decoder_block_size:]

            B, t = cur_dec.size()
            dec_pad_mask = torch.ones((B, t), dtype=torch.float,
                                      device=encoder_input_ids.device)

            logits, attn_weights, p_gen = self.decoder(
                cur_dec, encoder_output,
                decoder_pad_mask=dec_pad_mask,
                encoder_pad_mask=encoder_pad_mask
            )

            gen_probs  = F.softmax(logits, dim=-1)
            copy_probs = self._compute_copy_logits(attn_weights, encoder_input_ids)
            final_probs = p_gen * gen_probs + (1 - p_gen) * copy_probs   # (B, T, V)

            #convert to log space, apply temperature, re-softmax
            last_log_probs = torch.log(final_probs[:, -1, :].clamp(min=1e-10))

            # Repetition penalty
            if repetition_penalty != 1.0 and decoder_input_ids.size(1) > 1:
                for tok_id in decoder_input_ids[0].tolist():
                    last_log_probs[:, tok_id] /= repetition_penalty

            last_log_probs = last_log_probs / temperature    # temperature scaling


            if top_k > 0:
                topk_vals, _ = torch.topk(last_log_probs, top_k)
                last_log_probs[last_log_probs < topk_vals[:, [-1]]] = float('-inf')

            probs = F.softmax(last_log_probs, dim=-1)

            next_token = torch.multinomial(probs, num_samples=1)
            decoder_input_ids = torch.cat([decoder_input_ids, next_token], dim=1)

            if eos_token_id is not None and next_token.item() == eos_token_id:
                break

        return decoder_input_ids


print("Fixed model loaded")
```

```python
# ===============================================
#  TRAINING LOOP — 25 EPOCH FULL RUN (RESUMABLE)
# ===============================================

import json
import torch
import joblib
from torch.utils.data import Dataset, DataLoader
from transformers import GPT2TokenizerFast
from tqdm import tqdm
from google.colab import drive
import os
import torch.nn.functional as F
import numpy as np

# -------- Drive --------
drive.mount("/content/drive", force_remount=False)

# -------- Device --------
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# -------- Paths --------
DRIVE_BASE_PATH  = "/content/drive/MyDrive/readme_training1"
DATA_PATH        = f"{DRIVE_BASE_PATH}/summaries_final.json"
MODEL_SAVE_PATH  = f"{DRIVE_BASE_PATH}/readme_seq2seq_model_state.pt"
TOKENIZER_SAVE_PATH = f"{DRIVE_BASE_PATH}/readme_seq2seq_tokenizer.joblib"

os.makedirs(DRIVE_BASE_PATH, exist_ok=True)

# -------- (NEW) Resume from checkpoint --------
# If you want to resume, set this to your checkpoint file.
# If you want a fresh run, set it to None.
RESUME_CKPT_PATH = f"{DRIVE_BASE_PATH}/readme_seq2seq_model_state_epoch24.pt"
# RESUME_CKPT_PATH = None

# -------- Clear cache --------
print("Clearing GPU cache...")
if 'model' in globals() and isinstance(model, torch.nn.Module):
    del model
if torch.cuda.is_available():
    torch.cuda.empty_cache()
print("✅ Cache cleared.")

# -------- Tokenizer --------
tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")
tokenizer.add_special_tokens({
    "pad_token": "<|pad|>",
    "additional_special_tokens": ["<|sep|>", "<|eos|>"],
})

PAD_ID = tokenizer.pad_token_id
SEP_ID = tokenizer.convert_tokens_to_ids("<|sep|>")
EOS_ID = tokenizer.convert_tokens_to_ids("<|eos|>")

print(f"✅ Tokenizer: PAD={PAD_ID}, SEP(BOS)={SEP_ID}, EOS={EOS_ID}")

# -------- Config --------
cfg = config(
    vocab_size=len(tokenizer),
    encoder_block_size=1024,
    decoder_block_size=256,
    n_layer=4,
    n_head=8,
    n_embd=320,
    dropout=0.1,
    pad_token_id=PAD_ID,
    special_token_ids=[SEP_ID, EOS_ID],   # copy mechanism never outputs control tokens
)

print(f"✅ Config: {cfg}")

# -------- Model --------
model = Seq2SeqModel(cfg).to(DEVICE)

# Resize embeddings to match tokenizer (3 new special tokens added)
model.encoder.wte    = torch.nn.Embedding(len(tokenizer), model.config.n_embd).to(DEVICE, dtype=torch.float32)
model.decoder.wte    = torch.nn.Embedding(len(tokenizer), model.config.n_embd).to(DEVICE, dtype=torch.float32)
model.decoder.lm_head.weight = model.decoder.wte.weight   # re-tie after resize

print(f"✅ Model: {sum(p.numel() for p in model.parameters())/1e6:.2f}M params")


# ════════════════════════════════════════════════════════════════════════════
#  DATASET
# ════════════════════════════════════════════════════════════════════════════

class ReadmeSummaryDataset(Dataset):
    def __init__(self, path, tokenizer, encoder_block_size, decoder_block_size,
                 min_summary_tokens=10):
        self.encoder_block_size = encoder_block_size
        self.decoder_block_size = decoder_block_size

        if not os.path.exists(path):
            raise FileNotFoundError(f"Data file not found: {path}")

        with open(path, "r") as f:
            raw_data = json.load(f)

        self.tokenizer = tokenizer
        self.data      = []

        for item in raw_data:
            readme  = item.get("readme",  "").strip()
            summary = item.get("summary", "").strip()
            if not readme or not summary:
                continue

            readme_ids  = tokenizer.encode(readme)
            summary_ids = tokenizer.encode(summary)

            if len(summary_ids) < min_summary_tokens:
                continue
            if len(readme_ids) > encoder_block_size:
                continue
            if (1 + len(summary_ids)) > decoder_block_size:
                continue

            self.data.append({"readme_ids": readme_ids, "summary_ids": summary_ids})

        print(f"✅ Dataset: {len(self.data)} valid samples")

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]

        # Encoder input: readme tokens, right-padded
        enc = item["readme_ids"]
        enc = enc + [self.tokenizer.pad_token_id] * (self.encoder_block_size - len(enc))

        # Decoder input:  [SEP, tok1, tok2, ...], right-padded
        dec_in = [SEP_ID] + item["summary_ids"]
        dec_in = dec_in + [self.tokenizer.pad_token_id] * (self.decoder_block_size - len(dec_in))

        # Decoder target: [tok1, tok2, ..., EOS], -100 padded (ignored by loss)
        dec_tgt = item["summary_ids"] + [EOS_ID]
        dec_tgt = dec_tgt + [-100] * (self.decoder_block_size - len(dec_tgt))

        return (
            torch.tensor(enc,     dtype=torch.long),
            torch.tensor(dec_in,  dtype=torch.long),
            torch.tensor(dec_tgt, dtype=torch.long),
        )


# ════════════════════════════════════════════════════════════════════════════
#  DATALOADER / OPTIMIZER / SCHEDULER
# ════════════════════════════════════════════════════════════════════════════

dataset = ReadmeSummaryDataset(
    DATA_PATH, tokenizer,
    cfg.encoder_block_size,
    cfg.decoder_block_size
)
loader = DataLoader(
    dataset,
    batch_size=8,
    shuffle=True,
    pin_memory=(DEVICE == 'cuda')
)

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=5e-4,
    betas=(0.9, 0.95),
    weight_decay=0.1
)

from torch.optim.lr_scheduler import OneCycleLR

EPOCHS              = 25
GRAD_CLIP           = 1.0
DIAGNOSTIC_INTERVAL = 100

scheduler = OneCycleLR(
    optimizer,
    max_lr=5e-4,
    epochs=EPOCHS,
    steps_per_epoch=len(loader),
    pct_start=0.1,
    anneal_strategy='cos'
)

# FIX: non-deprecated GradScaler API
scaler = torch.amp.GradScaler('cuda')

# ════════════════════════════════════════════════════════════════════════════
#  (NEW) RESUME LOGIC
# ════════════════════════════════════════════════════════════════════════════

start_epoch = 0
best_nll       = float('inf')
prev_ckpt_path = None

if RESUME_CKPT_PATH is not None:
    if not os.path.exists(RESUME_CKPT_PATH):
        raise FileNotFoundError(f"Resume checkpoint not found: {RESUME_CKPT_PATH}")

    ckpt = torch.load(RESUME_CKPT_PATH, map_location="cpu")

    model.load_state_dict(ckpt["model_state_dict"])
    optimizer.load_state_dict(ckpt["optimizer_state_dict"])

    # saved as 1-based epoch index (epoch+1)
    start_epoch = int(ckpt.get("epoch", 0))
    print(f"✅ Resumed from: {os.path.basename(RESUME_CKPT_PATH)} (completed epoch {start_epoch})")

    # Continue rolling checkpoint deletion from the resume file
    prev_ckpt_path = RESUME_CKPT_PATH

    # Rebuild scheduler and advance it to the correct step (best-effort)
    steps_per_epoch = len(loader)
    total_steps_to_advance = start_epoch * steps_per_epoch

    scheduler = OneCycleLR(
        optimizer,
        max_lr=5e-4,
        epochs=EPOCHS,
        steps_per_epoch=steps_per_epoch,
        pct_start=0.1,
        anneal_strategy='cos'
    )

    for _ in range(total_steps_to_advance):
        scheduler.step()

    print(f"✅ Scheduler advanced by {total_steps_to_advance} steps")

    # If metrics exist, use them for a reasonable "best so far" display
    if isinstance(ckpt.get("metrics", None), dict) and "nll" in ckpt["metrics"]:
        try:
            best_nll = float(ckpt["metrics"]["nll"])
        except Exception:
            pass


# ════════════════════════════════════════════════════════════════════════════
#  HELPERS
# ════════════════════════════════════════════════════════════════════════════

def get_sampling_prob(epoch, total_epochs):
    """
    Don't start scheduled sampling until epoch 10.
    At epoch 5 the model was still producing garbage — feeding that back in
    as training inputs just adds noise.  Ramp 0 → 0.3 over epochs 10-25.
    Capped at 0.3 (not 0.5) because 25% was already hurting NLL at epoch 5.
    """
    if epoch < 10:
        return 0.0
    return min(0.3, ((epoch - 10) / max(total_epochs - 10, 1)) * 0.3)


def run_inference(model, tokenizer, readme_text, device,
                  max_new_tokens=120, temperature=None, top_k=40, epoch=0):
    """
    Adaptive temperature:
      epochs  0-9:  0.5  — forces commitment to most confident tokens,
                           good for spotting signal in an undertrained model
      epochs 10-24: 0.8  — more natural variation once the model has learned something
    """
    if temperature is None:
        temperature = 0.5 if epoch < 10 else 0.8

    model.eval()
    enc_ids = tokenizer.encode(readme_text)
    if len(enc_ids) > model.config.encoder_block_size:
        enc_ids = enc_ids[:model.config.encoder_block_size]

    encoder_input_ids = torch.tensor([enc_ids], dtype=torch.long, device=device)

    output_ids = model.summarize(
        encoder_input_ids,
        max_new_tokens=max_new_tokens,
        temperature=temperature,
        top_k=top_k,
        bos_token_id=SEP_ID,
        eos_token_id=EOS_ID,
        repetition_penalty=1.2,
    )

    generated = output_ids[0, 1:].tolist()
    if EOS_ID in generated:
        generated = generated[:generated.index(EOS_ID)]

    model.train()
    return tokenizer.decode(generated, skip_special_tokens=True)


def run_inference_block(model, tokenizer, dataset, device, epoch, n_samples=3):
    """
    Run inference on samples spread across the dataset.
    Prints each generated summary alongside the reference so you can
    compare directly without having to run separate code.
    """
    print(f"\n{'━'*70}")
    print(f"🔍 INFERENCE CHECK — Epoch {epoch+1}  "
          f"(temp={'0.5 [low — early model]' if epoch < 10 else '0.8'})")
    print(f"{'━'*70}")

    indices = [0, len(dataset)//2, len(dataset)-1][:n_samples]

    for i, idx in enumerate(indices):
        item        = dataset.data[idx]
        readme_text = tokenizer.decode(item["readme_ids"])
        reference   = tokenizer.decode(item["summary_ids"])
        generated   = run_inference(model, tokenizer, readme_text, device, epoch=epoch)

        readme_preview = readme_text[:200].replace('\n', ' ')
        ref_preview    = reference[:200]
        gen_preview    = generated[:200]

        print(f"\n  Sample {i+1}  (dataset[{idx}])")
        print(f"  {'─'*66}")
        print(f"  README (first 200 chars):")
        print(f"    {readme_preview}...")
        print(f"  Reference:")
        print(f"    {ref_preview}")
        print(f"  Generated:")
        if gen_preview.strip():
            print(f"    {gen_preview}")
        else:
            print(f"    [empty output — model not ready yet, this is normal early on]")

    print(f"\n{'━'*70}\n")


# ════════════════════════════════════════════════════════════════════════════
#  CHECKPOINT MANAGEMENT
#  - Rolling checkpoint: saved every epoch, previous deleted → Drive stays lean
#  - Best NLL checkpoint: never deleted, always reflects the best weights seen
#  - Final checkpoint: saved at the very end under the canonical filename
# ════════════════════════════════════════════════════════════════════════════

def save_checkpoint(model, optimizer, epoch, avg_metrics, prev_path=None):
    """Save current epoch, delete previous rolling checkpoint."""
    path = MODEL_SAVE_PATH.replace('.pt', f'_epoch{epoch+1}.pt')
    torch.save({
        'epoch':                epoch + 1,
        'model_state_dict':     model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'metrics':              avg_metrics,
    }, path)
    print(f"  💾 Saved  → {os.path.basename(path)}")

    if prev_path and os.path.exists(prev_path) and prev_path != path:
        os.remove(prev_path)
        print(f"  🗑️  Removed → {os.path.basename(prev_path)}")

    return path


def maybe_save_best(model, optimizer, epoch, avg_nll, best_nll):
    """Overwrite best-NLL checkpoint if this epoch improves it."""
    if avg_nll < best_nll:
        best_path = MODEL_SAVE_PATH.replace('.pt', '_best_nll.pt')
        torch.save({
            'epoch':                epoch + 1,
            'model_state_dict':     model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'avg_nll':              avg_nll,
        }, best_path)
        print(f"  🏆 New best NLL {avg_nll:.4f} → {os.path.basename(best_path)}")
        return avg_nll
    return best_nll


# ════════════════════════════════════════════════════════════════════════════
#  TRAINING LOOP
# ════════════════════════════════════════════════════════════════════════════

model.train()

print(f"\n🚀 Training for {EPOCHS} epochs on {len(dataset)} samples")
print(f"   Batch size: 8  |  Steps/epoch: {len(loader)}")
print(f"   Scheduled sampling: off until epoch 10, then 0 → 30%")
print(f"   Inference check: every 5 epochs (3 samples with reference)")
print(f"   Checkpoints: every epoch rolling + permanent best-NLL")
if start_epoch > 0:
    print(f"   ✅ Resuming at epoch {start_epoch+1}/{EPOCHS} from checkpoint\n")
else:
    print()

for epoch in range(start_epoch, EPOCHS):

    sampling_prob          = get_sampling_prob(epoch, EPOCHS)
    use_scheduled_sampling = (epoch >= 10)

    pbar = tqdm(loader, desc=f"Epoch {epoch+1}/{EPOCHS}")

    epoch_metrics = {
        'nll':             [],
        'attn_entropy':    [],
        'mean_cross_attn': [],
        'p_gen':           [],
        'total_loss':      [],
    }

    for batch_idx, (enc_ids, dec_ids, tgt_ids) in enumerate(pbar):

        enc_ids = enc_ids.to(DEVICE)
        dec_ids = dec_ids.to(DEVICE)
        tgt_ids = tgt_ids.to(DEVICE)

        optimizer.zero_grad(set_to_none=True)

        # FIX: non-deprecated autocast API
        with torch.amp.autocast('cuda', enabled=True, dtype=torch.float16):
            final_log_probs, loss, metrics = model(
                enc_ids, dec_ids, tgt_ids,
                use_scheduled_sampling=use_scheduled_sampling,
                sampling_prob=sampling_prob,
            )

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), GRAD_CLIP)
        scaler.step(optimizer)
        scaler.update()
        scheduler.step()

        epoch_metrics['nll'].append(metrics['nll_loss'])
        epoch_metrics['attn_entropy'].append(metrics['attn_entropy'])
        epoch_metrics['mean_cross_attn'].append(metrics['mean_cross_attn'])
        epoch_metrics['p_gen'].append(metrics['p_gen_mean'])
        epoch_metrics['total_loss'].append(loss.item())

        current_lr = scheduler.get_last_lr()[0]
        pbar.set_postfix(
            loss  =f"{loss.item():.4f}",
            nll   =f"{metrics['nll_loss']:.4f}",
            p_gen =f"{metrics['p_gen_mean']:.3f}",
            attn_e=f"{metrics['attn_entropy']:.2f}",
            lr    =f"{current_lr:.2e}",
        )

        # ── Mid-epoch diagnostics ──────────────────────────────────────────
        if batch_idx % DIAGNOSTIC_INTERVAL == 0 and batch_idx > 0:
            print(f"\n{'='*70}")
            print(f"📊 DIAGNOSTICS — Epoch {epoch+1}, Batch {batch_idx+1}")
            print(f"{'='*70}")
            print(f"  Total Loss:           {loss.item():.4f}")
            print(f"  NLL Loss:             {metrics['nll_loss']:.4f}"
                  f"  {'✅' if metrics['nll_loss'] < 7.0 else '  📉 still learning'}")
            print(f"  p_gen (mean):         {metrics['p_gen_mean']:.3f}  ", end="")
            if metrics['p_gen_mean'] > 0.92:
                print("⚠️  very high — copy barely used")
            elif metrics['p_gen_mean'] < 0.3:
                print("⚠️  very low — over-copying")
            elif 0.5 < metrics['p_gen_mean'] < 0.85:
                print("✅ good balance")
            else:
                print("📈 learning")

            print(f"\n  🎯 Encoder Usage:")
            print(f"  Attn Entropy:         {metrics['attn_entropy']:.3f}  "
                  f"{'✅ focused' if metrics['attn_entropy'] < 3.0 else '⚠️  too uniform (ignoring encoder)'}")
            print(f"  Mean Cross-Attn:      {metrics['mean_cross_attn']:.3f}  "
                  f"{'✅ active' if metrics['mean_cross_attn'] > 0.5 else '🚨 encoder being ignored!'}")

            if metrics['mean_cross_attn'] < 0.3:
                print(f"  → Consider raising attention_usage_penalty weight in model forward().")

            if use_scheduled_sampling:
                print(f"\n  Scheduled Sampling:   {sampling_prob:.1%} of inputs replaced")

            print(f"{'='*70}\n")

    # ── Epoch summary ──────────────────────────────────────────────────────
    avg = {k: (sum(v) / len(v)) for k, v in epoch_metrics.items()}

    print(f"\n{'='*70}")
    print(f"📈 Epoch {epoch+1}/{EPOCHS} Summary")
    print(f"{'='*70}")
    print(f"  Avg Total Loss:       {avg['total_loss']:.4f}")
    print(f"  Avg NLL:              {avg['nll']:.4f}  (random = 10.82  |  good = <4.0)")
    print(f"  Avg p_gen:            {avg['p_gen']:.3f}")
    print(f"  Avg Attn Entropy:     {avg['attn_entropy']:.3f}")
    print(f"  Avg Cross-Attention:  {avg['mean_cross_attn']:.3f}")
    print(f"  LR end of epoch:      {current_lr:.2e}")
    print(f"  Best NLL so far:      {min(best_nll, avg['nll']):.4f}")

    # Health warnings
    if avg['mean_cross_attn'] < 0.4:
        print(f"\n  🚨 Low cross-attention — encoder may be underutilised.")
    if avg['nll'] > 9.5 and epoch > 4:
        print(f"\n  ⚠️  NLL still very high after epoch {epoch+1}. "
              f"Check data quality or try a lower learning rate.")
    if avg['p_gen'] > 0.92:
        print(f"\n  ⚠️  p_gen {avg['p_gen']:.3f} — copy mechanism barely active. "
              f"Consider reducing gen_encouragement weight.")
    if use_scheduled_sampling and avg['nll'] > avg.get('prev_nll', avg['nll']):
        print(f"\n  ⚠️  NLL went up after scheduled sampling started. "
              f"Model may not be ready — watch next epoch.")

    # ── Inference check every 5 epochs ────────────────────────────────────
    if (epoch + 1) % 5 == 0:
        run_inference_block(model, tokenizer, dataset, DEVICE, epoch, n_samples=3)

    print(f"{'='*70}\n")

    # ── Checkpoints ───────────────────────────────────────────────────────
    prev_ckpt_path = save_checkpoint(model, optimizer, epoch, avg, prev_ckpt_path)
    best_nll       = maybe_save_best(model, optimizer, epoch, avg['nll'], best_nll)

    print()  # spacing


# ════════════════════════════════════════════════════════════════════════════
#  DONE
# ════════════════════════════════════════════════════════════════════════════

print("\n✅ Training complete!")
print(f"   Best NLL: {best_nll:.4f}")

torch.save({
    'epoch':                EPOCHS,
    'model_state_dict':     model.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
    'config':               cfg,
    'final_metrics':        avg,
    'best_nll':             best_nll,
}, MODEL_SAVE_PATH)

joblib.dump(tokenizer, TOKENIZER_SAVE_PATH)
print(f"  💾 Final model  → {MODEL_SAVE_PATH}")
print(f"  💾 Tokenizer    → {TOKENIZER_SAVE_PATH}")

# Final inference
print("\n" + "═"*70)
print("🏁 FINAL INFERENCE TEST — loading best NLL checkpoint")
print("═"*70)

best_path = MODEL_SAVE_PATH.replace('.pt', '_best_nll.pt')
if not os.path.exists(best_path):
    print(f"⚠️ Best checkpoint not found yet: {best_path}")
else:
    best_ckpt = torch.load(best_path, map_location="cpu")
    model.load_state_dict(best_ckpt['model_state_dict'])
    print(f"   Loaded epoch {best_ckpt['epoch']} (NLL {best_ckpt['avg_nll']:.4f})")
    run_inference_block(model, tokenizer, dataset, DEVICE, epoch=EPOCHS-1, n_samples=3)
```

```python
# ===============================================
#  README SUMMARIZER - TESTING INTERFACE
# ===============================================

import torch
import joblib
import json
from google.colab import drive
from transformers import GPT2TokenizerFast
import os
from dataclasses import dataclass, field
from typing import List
import random

# -------- Mount Drive --------
print("📁 Mounting Google Drive...")
drive.mount('/content/drive', force_remount=False)
print("✅ Drive mounted!\n")

# -------- Paths --------
DRIVE_BASE_PATH     = "/content/drive/MyDrive/readme_training1"
MODEL_PATH          = f"{DRIVE_BASE_PATH}/readme_seq2seq_model_state.pt"
BEST_NLL_PATH       = f"{DRIVE_BASE_PATH}/readme_seq2seq_model_state_best_nll.pt"
TOKENIZER_PATH      = f"{DRIVE_BASE_PATH}/readme_seq2seq_tokenizer.joblib"
DATA_PATH           = f"{DRIVE_BASE_PATH}/summaries_final.json"
RESULTS_PATH        = f"{DRIVE_BASE_PATH}/test_results.json"

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print(f"🖥️  Device: {DEVICE}")
if DEVICE == "cuda":
    print(f"   GPU: {torch.cuda.get_device_name(0)}")
    print(f"   Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB\n")

# -------- Config (must match training) --------
# FIX: added special_token_ids field that the copy mechanism requires
@dataclass
class config:
    vocab_size: int
    encoder_block_size: int
    decoder_block_size: int
    n_layer: int
    n_head: int
    n_embd: int
    dropout: float
    pad_token_id: int
    special_token_ids: List[int] = field(default_factory=list)

print("⚠️  Make sure model architecture classes are loaded before running!\n")

# -------- Check files --------
print("🔍 Checking for saved files...")
for path in [TOKENIZER_PATH]:
    if os.path.exists(path):
        print(f"   ✅ {os.path.basename(path)} ({os.path.getsize(path)/1024**2:.1f} MB)")
    else:
        print(f"   ❌ {os.path.basename(path)} NOT FOUND")
        raise FileNotFoundError(f"Missing: {path}")

# Pick best available model checkpoint
# Priority: best_nll > final > any epoch checkpoint
def find_best_checkpoint():
    if os.path.exists(BEST_NLL_PATH):
        return BEST_NLL_PATH, "best NLL"
    if os.path.exists(MODEL_PATH):
        return MODEL_PATH, "final"
    # scan for epoch checkpoints
    candidates = sorted([
        f for f in os.listdir(DRIVE_BASE_PATH)
        if f.startswith("readme_seq2seq_model_state_epoch") and f.endswith(".pt")
    ], key=lambda x: int(x.split("epoch")[1].split(".")[0]))
    if candidates:
        path = f"{DRIVE_BASE_PATH}/{candidates[-1]}"
        return path, f"epoch checkpoint ({candidates[-1]})"
    raise FileNotFoundError("No model checkpoint found in Drive.")

LOAD_PATH, LOAD_SOURCE = find_best_checkpoint()
print(f"   ✅ Model checkpoint: {os.path.basename(LOAD_PATH)} ({LOAD_SOURCE})")
print(f"      Size: {os.path.getsize(LOAD_PATH)/1024**2:.1f} MB\n")

# -------- Load tokenizer --------
print("📚 Loading tokenizer...")
tokenizer = joblib.load(TOKENIZER_PATH)

PAD_ID = tokenizer.pad_token_id
SEP_ID = tokenizer.convert_tokens_to_ids("<|sep|>")
EOS_ID = tokenizer.convert_tokens_to_ids("<|eos|>")

print(f"✅ Tokenizer loaded — vocab: {len(tokenizer)}, PAD={PAD_ID}, SEP={SEP_ID}, EOS={EOS_ID}\n")

# -------- Load model --------
print(f"🤖 Loading model from {LOAD_SOURCE}...")

cfg = config(
    vocab_size=len(tokenizer),
    encoder_block_size=1024,
    decoder_block_size=256,
    n_layer=4,
    n_head=8,
    n_embd=320,
    dropout=0.1,
    pad_token_id=PAD_ID,
    special_token_ids=[SEP_ID, EOS_ID],   # FIX: required by copy mechanism
)

model = Seq2SeqModel(cfg).to(DEVICE)

checkpoint = torch.load(LOAD_PATH, map_location=DEVICE)
if isinstance(checkpoint, dict) and 'model_state_dict' in checkpoint:
    model.load_state_dict(checkpoint['model_state_dict'])
    print(f"✅ Loaded checkpoint (epoch {checkpoint.get('epoch', '?')})")
    # print training metrics if available
    metrics = checkpoint.get('metrics') or checkpoint.get('final_metrics')
    if metrics:
        nll = metrics.get('nll', metrics.get('avg_nll', '?'))
        p_gen = metrics.get('p_gen', '?')
        print(f"   NLL at save: {nll:.4f}" if isinstance(nll, float) else f"   NLL: {nll}")
        print(f"   p_gen:       {p_gen:.3f}" if isinstance(p_gen, float) else f"   p_gen: {p_gen}")
    if 'avg_nll' in checkpoint:
        print(f"   Best NLL:    {checkpoint['avg_nll']:.4f}")
else:
    model.load_state_dict(checkpoint)
    print("✅ Loaded state dict directly")

model.eval()
print(f"   Parameters: {sum(p.numel() for p in model.parameters())/1e6:.2f}M\n")


# ════════════════════════════════════════════════════════════════════════════
#  CORE INFERENCE
# ════════════════════════════════════════════════════════════════════════════

def summarize_text(readme_text, max_tokens=80, temperature=0.7, top_k=40,
                   repetition_penalty=1.2, show_diagnostics=False): #for prod im thinkng 50 max tokens and 1.5 repitition pentanly to fix issues
    """
    Generate summary for a README.

    Args:
        readme_text:        Input README string
        max_tokens:         Max tokens to generate (default 80, was 50)
        temperature:        Lower = more focused (0.3-0.5 early, 0.7-0.8 trained)
        top_k:              Top-k sampling
        repetition_penalty: >1.0 penalises repeated tokens (fixes "and and and")
        show_diagnostics:   Print p_gen / attention stats
    """
    enc_ids = tokenizer.encode(readme_text)
    # FIX: truncate long inputs instead of crashing
    if len(enc_ids) > cfg.encoder_block_size:
        enc_ids = enc_ids[:cfg.encoder_block_size]

    encoder_input_ids = torch.tensor([enc_ids], dtype=torch.long, device=DEVICE)

    with torch.no_grad():
        generated_ids = model.summarize(
            encoder_input_ids,
            max_new_tokens=max_tokens,
            temperature=temperature,
            top_k=top_k,
            bos_token_id=SEP_ID,
            eos_token_id=EOS_ID,
            repetition_penalty=repetition_penalty,   # FIX: was missing entirely
        )

    # strip BOS and EOS
    generated = generated_ids[0, 1:].tolist()
    if EOS_ID in generated:
        generated = generated[:generated.index(EOS_ID)]
    summary = tokenizer.decode(generated, skip_special_tokens=True)

    if show_diagnostics:
        with torch.no_grad():
            encoder_pad_mask = (encoder_input_ids != PAD_ID).float()
            encoder_output   = model.encoder(encoder_input_ids, pad_mask=encoder_pad_mask)

            dec_ids      = generated_ids
            dec_pad_mask = torch.ones(dec_ids.shape, dtype=torch.float, device=DEVICE)

            logits, attn_weights, p_gen = model.decoder(
                dec_ids, encoder_output,
                decoder_pad_mask=dec_pad_mask,
                encoder_pad_mask=encoder_pad_mask
            )

            p_gen_vals = p_gen[0].squeeze(-1).cpu().numpy()
            attn_mean  = attn_weights[0].mean(dim=0).cpu().numpy()

            print(f"\n  📊 Diagnostics:")
            print(f"     p_gen mean / min / max: "
                  f"{p_gen_vals.mean():.3f} / {p_gen_vals.min():.3f} / {p_gen_vals.max():.3f}")
            print(f"     {'Mostly generating' if p_gen_vals.mean() > 0.6 else 'Mostly copying'}")
            print(f"     Attn mean / max: {attn_mean.mean():.3f} / {attn_mean.max():.3f}")
            entropy = -(attn_mean * (attn_mean + 1e-10)).sum()
            print(f"     Attn entropy:    {abs(entropy):.3f}  "
                  f"({'focused' if abs(entropy) < 3.0 else 'spread'})")

    return summary


# ════════════════════════════════════════════════════════════════════════════
#  METRICS
# ════════════════════════════════════════════════════════════════════════════

def word_overlap(a, b):
    """Simple token-level F1 between two strings."""
    a_words = set(a.lower().split())
    b_words = set(b.lower().split())
    if not a_words or not b_words:
        return 0.0
    precision = len(a_words & b_words) / len(a_words)
    recall    = len(a_words & b_words) / len(b_words)
    if precision + recall == 0:
        return 0.0
    return 2 * precision * recall / (precision + recall)   # F1


def rouge_1(generated, reference):
    """Unigram ROUGE-1 F1 (no external library needed)."""
    return word_overlap(generated, reference)


def load_test_samples(n=10, seed=None):
    if not os.path.exists(DATA_PATH):
        print("❌ Training data not found — can't load samples")
        return []
    with open(DATA_PATH, 'r') as f:
        data = json.load(f)
    if seed is not None:
        random.seed(seed)
    return random.sample(data, min(n, len(data)))


# ════════════════════════════════════════════════════════════════════════════
#  DISPLAY HELPERS
# ════════════════════════════════════════════════════════════════════════════

def compare_summaries(readme, ground_truth=None, **kwargs):
    print("─" * 70)
    print("📝 README (first 400 chars):")
    print(f"   {readme[:400].replace(chr(10), ' ')}{'...' if len(readme) > 400 else ''}")

    summary = summarize_text(readme, **kwargs)

    print("\n🤖 Generated:")
    print(f"   {summary if summary.strip() else '[empty — model may need more training]'}")

    if ground_truth:
        print("\n🎯 Reference:")
        print(f"   {ground_truth[:300]}")
        score = rouge_1(summary, ground_truth)
        overlap_src = word_overlap(summary, readme)
        print(f"\n   ROUGE-1 F1 vs reference: {score*100:.1f}%")
        print(f"   Word overlap with source: {overlap_src*100:.1f}%  "
              f"({'high copy' if overlap_src > 0.6 else 'generating'})")

    print("─" * 70 + "\n")
    return summary


# ════════════════════════════════════════════════════════════════════════════
#  MODES
# ════════════════════════════════════════════════════════════════════════════

def mode_1_custom_input():
    print("\n📝 MODE 1: Custom README Input")
    print("Paste your README (blank line = use default):\n")

    default = (
        "FastAPI-based REST API framework with async support, automatic "
        "documentation generation, and built-in validation using Pydantic. "
        "Designed for high performance and developer productivity."
    )

    lines = []
    try:
        while True:
            line = input()
            if line == "" and not lines:
                break
            lines.append(line)
    except EOFError:
        pass

    readme = "\n".join(lines).strip() or default
    if not lines:
        print(f"(using default)\n{default}\n")

    try:
        max_tokens = int(input("Max tokens [80]: ").strip() or "80")
        temperature = float(input("Temperature [0.7]: ").strip() or "0.7")
        rep_pen = float(input("Repetition penalty [1.2]: ").strip() or "1.2")
        diag = input("Show diagnostics? [y/N]: ").strip().lower() == 'y'
    except Exception:
        max_tokens, temperature, rep_pen, diag = 80, 0.7, 1.2, False

    compare_summaries(readme, max_tokens=max_tokens, temperature=temperature,
                      repetition_penalty=rep_pen, show_diagnostics=diag)


def mode_2_test_samples():
    print("\n🔬 MODE 2: Test with Training Data Samples")

    try:
        n = int(input("How many samples? [5]: ").strip() or "5")
    except Exception:
        n = 5

    samples = load_test_samples(n=n)
    if not samples:
        return

    for i, sample in enumerate(samples, 1):
        print(f"\nSAMPLE {i}/{len(samples)} — {sample.get('repo_name', 'unknown')}")
        compare_summaries(
            sample.get('readme', ''),
            ground_truth=sample.get('summary'),
            temperature=0.7,
            repetition_penalty=1.2,
            show_diagnostics=True,
        )
        if i < len(samples):
            if input("Next? [Enter / q]: ").strip().lower() == 'q':
                break


def mode_3_batch_test():
    print("\n📦 MODE 3: Batch Test")

    try:
        n = int(input("How many samples? [20]: ").strip() or "20")
    except Exception:
        n = 20

    samples = load_test_samples(n=n, seed=42)   # fixed seed for reproducibility
    if not samples:
        return

    results = []
    rouge_scores   = []
    src_overlaps   = []
    lengths        = []

    for i, sample in enumerate(samples, 1):
        readme = sample.get('readme', '')
        gt     = sample.get('summary', '')

        print(f"  [{i:>2}/{n}] {sample.get('repo_name', '?')[:40]:<40}", end=' ')

        summary = summarize_text(readme, temperature=0.7, repetition_penalty=1.2)

        r1          = rouge_1(summary, gt)
        src_overlap = word_overlap(summary, readme)
        length      = len(summary.split())

        rouge_scores.append(r1)
        src_overlaps.append(src_overlap)
        lengths.append(length)

        print(f"ROUGE-1: {r1*100:.0f}%  src_overlap: {src_overlap*100:.0f}%  len: {length}")

        results.append({
            'repo_name':          sample.get('repo_name', '?'),
            'generated':          summary,
            'ground_truth':       gt,
            'rouge_1':            r1,
            'source_overlap':     src_overlap,
            'summary_length':     length,
        })

    print(f"\n{'='*70}")
    print(f"📊 BATCH RESULTS ({n} samples)")
    print(f"{'='*70}")
    print(f"  Avg ROUGE-1 F1:          {sum(rouge_scores)/len(rouge_scores)*100:.1f}%")
    print(f"  Avg source word overlap: {sum(src_overlaps)/len(src_overlaps)*100:.1f}%")
    print(f"  Avg summary length:      {sum(lengths)/len(lengths):.1f} words")
    print(f"\n  ROUGE-1 interpretation:")
    avg_r1 = sum(rouge_scores)/len(rouge_scores)
    if avg_r1 > 0.4:
        print(f"  ✅ Strong overlap with reference summaries")
    elif avg_r1 > 0.2:
        print(f"  📈 Moderate overlap — model has learned the topic but not the phrasing")
    else:
        print(f"  📉 Low overlap — model generating different content than reference")

    print(f"\n  Source overlap interpretation:")
    avg_src = sum(src_overlaps)/len(src_overlaps)
    if avg_src > 0.7:
        print(f"  ⚠️  Very high ({avg_src*100:.0f}%) — model may be over-copying from source")
    elif avg_src > 0.4:
        print(f"  ✅ Good ({avg_src*100:.0f}%) — healthy mix of copy and generation")
    else:
        print(f"  📈 Low ({avg_src*100:.0f}%) — model generating most content independently")

    # worst performers
    results_sorted = sorted(results, key=lambda x: x['rouge_1'])
    print(f"\n  Worst 3 (by ROUGE-1):")
    for r in results_sorted[:3]:
        print(f"    {r['repo_name'][:35]:<35} ROUGE-1: {r['rouge_1']*100:.0f}%")
        print(f"      Gen: {r['generated'][:80]}")
        print(f"      Ref: {r['ground_truth'][:80]}")

    with open(RESULTS_PATH, 'w') as f:
        json.dump(results, f, indent=2)
    print(f"\n  💾 Results saved → {RESULTS_PATH}")


def mode_4_temperature_sweep():
    """Test the same README at different temperatures to find the sweet spot."""
    print("\n🌡️  MODE 4: Temperature Sweep")
    print("Tests one README at multiple temperatures so you can compare.\n")

    sample = load_test_samples(n=1, seed=7)
    if not sample:
        return
    sample = sample[0]

    readme = sample.get('readme', '')
    gt     = sample.get('summary', '')

    print(f"Repo: {sample.get('repo_name', '?')}")
    print(f"README: {readme[:200].replace(chr(10), ' ')}...\n")
    print(f"Reference: {gt[:200]}\n")
    print("─" * 70)

    for temp in [0.3, 0.5, 0.7, 0.9, 1.1]:
        summary = summarize_text(readme, temperature=temp, repetition_penalty=1.2)
        r1      = rouge_1(summary, gt)
        print(f"  temp={temp:.1f}  ROUGE-1={r1*100:.0f}%")
        print(f"    {summary}")
        print()


# ════════════════════════════════════════════════════════════════════════════
#  MAIN MENU
# ════════════════════════════════════════════════════════════════════════════

def main_menu():
    while True:
        print("\n" + "="*70)
        print("🧪 README SUMMARIZER — TESTING INTERFACE")
        print(f"   Loaded: {os.path.basename(LOAD_PATH)}  ({LOAD_SOURCE})")
        print("="*70)
        print("  1. Custom README input")
        print("  2. Test with training data samples (with reference)")
        print("  3. Batch test — ROUGE-1 scores on N samples")
        print("  4. Temperature sweep — find best temperature")
        print("  5. Exit")
        print("="*70)

        choice = input("\nChoice (1-5): ").strip()

        if choice == '1':   mode_1_custom_input()
        elif choice == '2': mode_2_test_samples()
        elif choice == '3': mode_3_batch_test()
        elif choice == '4': mode_4_temperature_sweep()
        elif choice == '5':
            print("\n👋 Done!")
            break
        else:
            print("❌ Enter 1-5")


# ════════════════════════════════════════════════════════════════════════════
#  AUTO QUICK TEST ON LOAD
# ════════════════════════════════════════════════════════════════════════════

print("="*70)
print("🎉 MODEL LOADED — running quick test...\n")

_quick_readme = """
This project is a lightweight web framework for Python that provides
routing, templating, and middleware support. It's designed for building
microservices and APIs with minimal overhead and maximum flexibility.
Supports async handlers, JWT authentication, and ships with a CLI tool
for scaffolding new projects.
"""

_summary = summarize_text(_quick_readme, max_tokens=60, temperature=0.7,
                           repetition_penalty=1.2, show_diagnostics=True)
print(f"\n🤖 Quick test output:\n   {_summary}\n")
print("="*70)
print("\n💡 Temperature guide:")
print("   0.3-0.5 — focused/deterministic (good for early epochs)")
print("   0.7     — balanced (recommended)")
print("   0.9-1.1 — more varied (may get creative or noisy)\n")

main_menu()
```
